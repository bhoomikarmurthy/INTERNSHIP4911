# Problem Statement:

Big Basket, a food delivery service operating in multiple countries, aims to improve the overall customer experience by analysing the ratings provided by its customers. The target variable in this problem is the rating given by the customers after they receive their food delivery. The company wants to understand the factors that influence the customers' ratings and identify the areas that need improvement.



The challenge for Big Basket is to collect and analyse the ratings data from multiple countries, as customer preferences and expectations may vary across different regions. Additionally, the company needs to develop a system that can capture the feedback from customers in a timely and efficient manner to ensure that the issues are addressed promptly.



Thus, the problem statement for Big Basket is to develop a data-driven approach to understand the factors influencing customer ratings across different countries, and to use this information to enhance the overall customer experience by addressing the areas that require improvement.

```python
import re
#A regular expression (also known as "regex" or "regexp") is a sequence of characters that forms a search pattern. 
#It is used to match and manipulate strings, based on a specified pattern.
invalid_chars = r'[?#!$%&*@^]' # Regular expression for invalid characters

# Iterate over the columns of the dataset
for col in df.columns:
    # Convert the column to string type
    df[col] = df[col].astype(str)
    # Check if the column contains any invalid characters
    #regex=True is a parameter in the contains() method of Pandas Series object that tells the method to treat the pattern as a regular expression. 
    #When set to True, the pattern is interpreted as a regular expression, while when set to False (default), the pattern is treated as a literal string.
    invalid_mask = df[col].str.contains(invalid_chars, regex=True)
    if invalid_mask.any():
        # Find and print the column name, the invalid characters, and their total count
        invalid_chars_found = df.loc[invalid_mask, col].str.findall(invalid_chars)
        invalid_chars_found = [char for sublist in invalid_chars_found for char in sublist]
        invalid_chars_found_count = len(invalid_chars_found)
        invalid_chars_found = set(invalid_chars_found)
        invalid_chars_found = ', '.join(invalid_chars_found)
        print(f"Invalid characters found in column '{col}': {invalid_chars_found}. Total count: {invalid_chars_found_count}")
```

# CONVERTING THE MIXED DATATYPES INTO NUMERIC AND IMPUTING NULL VALUES

```python
import numpy as np
# Impute the "?" and "&" values in the "Average Cost for two" column
df["Average Cost for two"] = df["Average Cost for two"].replace(["?", "*", "&", "nan"], np.nan)

df["Average Cost for two"] = pd.to_numeric(df["Average Cost for two"])
df["Average Cost for two"] = df["Average Cost for two"].fillna(df["Average Cost for two"].median())

# Impute the "*" values in the "Price range" column
df["Price range"] = df["Price range"].replace(["*", "nan"], np.nan)
#errors="coerce" is a parameter used in pandas to handle errors while converting a column to a numeric data type. 
#When this parameter is set to "coerce", it means that if there are any invalid values (e.g., non-numeric values like strings) in the column, pandas will set them to NaN (Not a Number).
df["Price range"] = pd.to_numeric(df["Price range"], errors="coerce")
df["Price range"] = df["Price range"].fillna(df["Price range"].median())


# Save the cleaned dataset
df.to_csv("merged_dataset1.csv", index=False)
```

# OUTLIERS TREATEMENT

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loop over each numeric column in the dataframe and create a boxplot
for column in df1.select_dtypes(include=['float64', 'int64']):
    plt.figure(figsize=(8, 1))
    sns.boxplot(x=df1[column])
    plt.show()
```

From the above graph we can say that almost all the columns have the outliers present in it.

```python
def cap_data(df1):
    for col in df1.columns:
        print("capping the", col)
        if df1[col].dtype == 'float64' or df1[col].dtype == 'int64':
            percentiles = df1[col].quantile([0.20, 0.80]).values
            df1.loc[df1[col] <= percentiles[0], col] = percentiles[0]
            df1.loc[df1[col] >= percentiles[1], col] = percentiles[1]
        else:
            df1[col] = df1[col]
    return df1


df2 = cap_data(df1)
```

The capping method is a technique used in outlier treatment to limit extreme values by replacing them with a less extreme value that falls within a pre-determined range. In this method, any data point that is below a specified minimum value or above a specified maximum value is replaced with the nearest value within the range.

```python
# Loop over each numeric column in the dataframe and create a boxplot
for column in df2.select_dtypes(include=['float64', 'int64']):
    plt.figure(figsize=(8, 1))
    sns.boxplot(x=df2[column])
    plt.show()
```

From the above graph we can say that we have treated almost all the outliers present in all the outliers columns.

# VISUALIZING THE CORRELATION MATRIX

```python
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

with open('cleaned_data.csv', 'r', encoding='ISO-8859-1') as f:
    reader = csv.reader(f)
    clean_data = [row for row in reader]

# Get the column names and remove them from the data
column_names = clean_data[0]
clean_data = clean_data[1:]

numeric_data = []
mapping = {}
counter = 1
column_index_mapping = {}

for idx, column_name in enumerate(column_names):
    #Adds an entry to column_index_mapping, mapping the column name to its index in the list.
    column_index_mapping[column_name] = idx
    #Adds an entry to mapping, mapping the column name to the current value of counter.
    mapping[column_name] = counter
    #Increments counter by 1.
    counter += 1

for row in clean_data:
    numeric_row = []
    for value in row:
        # If the value is a string and hasn't been mapped yet, map it to a new numerical value
        if isinstance(value, str) and value not in mapping:
            mapping[value] = counter
            counter += 1
        # Convert the value to a numerical value and add it to the list of values for this row
        try:
            numeric_row.append(float(value))
        except ValueError:
            numeric_row.append(mapping[value])
    numeric_data.append(numeric_row)

numeric_data = np.array(numeric_data, dtype=np.float64)

corr_matrix = np.corrcoef(numeric_data, rowvar=False)

fig, ax = plt.subplots(figsize=(40, 30))
sns.heatmap(corr_matrix, ax=ax, annot=True, fmt='.2f')

# Use column names instead of numbers for the x and y axis labels
ax.set_title('Heatmap of Correlation Matrix')
ax.set_xlabel('Variable')
ax.set_ylabel('Variable')
ax.set_xticklabels(column_names)
ax.set_yticklabels(column_names)

plt.show()
```

Highly correlated columns are represented with light colors and less correlated columns are represented with dark colors.
