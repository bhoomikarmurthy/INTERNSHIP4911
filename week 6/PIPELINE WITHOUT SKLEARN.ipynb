```python
import numpy as np
import pandas as pd

# Define the K-nearest neighbors algorithm
class KNNRegressor:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X = X
        self.y = y
        
    def predict(self, X):
        y_pred = np.zeros(len(X))
        for i in range(len(X)):
            distances = np.sqrt(np.sum((self.X - X[i])**2, axis=1))
            idx = np.argsort(distances)[:self.k]
            y_pred[i] = np.mean(self.y[idx])
        return y_pred
```
